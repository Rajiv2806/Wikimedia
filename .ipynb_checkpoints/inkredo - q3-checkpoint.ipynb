```python
import pandas as pd
from __future__ import division
dataset = pd.read_csv("C:\Users\Rajiv2806\Documents\Resumes\Inkredo Assesement\events_log.csv")
dataset.head()
```

```python
dataset['time_formatted'] = dataset.timestamp.astype(str)

for i in range(len(dataset)):
    dataset.at[i,'Day'] = dataset.time_formatted.loc[i][7:9]
    
dataset.head()
```

```python
uniq_sessions = list(pd.unique(dataset.session_id.loc[0:100])) #Needs to be changed for overall dataset
total_sessions = len(uniq_sessions)
print "there are in total,",len(uniq_sessions),"No. of sessions"

visitPage_ct = 0

for i in uniq_sessions[0:100]:  #Needs to be changed for overall dataset
    if "visitPage" in list(dataset[dataset.session_id == i]['action']):
        visitPage_ct +=1

zero_results_rate = round(((total_sessions - visitPage_ct)/total_sessions)*100,5)

print "Overall Zero results rate:",zero_results_rate
```

```python
def func2(g,d):
    uniq_sessions = list(pd.unique(dataset[(dataset.group == g)&(dataset.Day == d)].session_id))
    total_sessions = len(uniq_sessions)

    visitPage_ct = 0
 
    for i in uniq_sessions[0:5]:    #Adujested in the interest of time
        if "visitPage" in list(dataset[dataset.session_id == i]['action']):
            visitPage_ct +=1

    zero_results_rate = round(((total_sessions - visitPage_ct)/total_sessions)*100,2)
    
    print "Overall Zero results rate of: ",g,"Day: ",d,"is: ",zero_results_rate
    return zero_results_rate
```

```python
import time
start_time = time.ctime()
print "-------Start Time:---------",start_time
run_time = time.clock()

Q3_df = pd.DataFrame()

Q3_df.at[1,'ga'] = func2('a','01')
Q3_df.at[1,'gb'] = func2('b','01')

Q3_df.at[2,'ga'] = func2('a','02')
Q3_df.at[2,'gb'] = func2('b','02')

Q3_df.at[3,'ga'] = func2('a','03')
Q3_df.at[3,'gb'] = func2('b','03')

Q3_df.at[4,'ga'] = func2('a','04')
Q3_df.at[4,'gb'] = func2('b','04')

Q3_df.at[5,'ga'] = func2('a','05')
Q3_df.at[5,'gb'] = func2('b','05')

Q3_df.at[6,'ga'] = func2('a','06')
Q3_df.at[6,'gb'] = func2('b','06')

Q3_df.at[7,'ga'] = func2('a','07')
Q3_df.at[7,'gb'] = func2('b','07')

Q3_df.at[8,'ga'] = func2('a','08')
Q3_df.at[8,'gb'] = func2('b','08')

print "------Execution time:-------",(time.clock()-run_time)/60," Minutes."
```

```python
import matplotlib.pyplot as plt
pos = list(range(len(Q3_df))) ;width = 0.25 

fig, ax = plt.subplots(figsize=(10,5))

plt.bar(pos, Q3_df['ga'],width, alpha=0.5,color='#EE3224',label=Q3_df.index[0]) 

plt.bar([p + width for p in pos], Q3_df['gb'],width,alpha=0.5,color='#F78F1E', label=Q3_df.index[1]) 

ax.set_ylabel('Result Position'); ax.set_xlabel('Day')

ax.set_title('Day-1 Result Positions for groups')

ax.set_xticks([p + 1.5 * width for p in pos])

ax.set_xticklabels(Q3_df.index)

plt.ylim([99, 100] )

plt.legend(['a', 'b'], loc='upper left')

plt.grid(); plt.show()
```
